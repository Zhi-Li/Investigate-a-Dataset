```python
## Calculate Correlation Coefficiency（Person`s R） 

## Survived values in std units
std_survived = titanic_df['Survived'] - titanic_df['Survived'].mean() / titanic_df['Survived'].std()

## Survived  female values in std units
## Note: Here I convert Sex into numerical data, where female is 1 and male is 0.
titanic_df['Sex'].iloc[0]

temp = []
for i in range(len(titanic_df)):
    if titanic_df['Sex'].iloc[i] == 'female':
        temp.append(1)
    else:
        temp.append(0)       
survived_female = pd.Series(temp)
std_female = survived_female - survived_female.mean() / survived_female.std()

r = (std_survived * std_female).mean()
print 'Person`s R: ', r
```

```python
## Now, I will calculate each embarked ports conditional proability repectively.
## Let`s set A as the event that passengers survived.And Bc/Bq/Bs represent passengers 
## who aboard at  (C = Cherbourg; Q = Queenstown; S = Southampton).
## Symbols: or -- ∪ ； and -- ∩ ; _ -- ( )

def count_Px(embarked):
    x = float(len(titanic_df[titanic_df.Embarked==embarked]))
    y = float(len(titanic_df[titanic_df.Embarked != 'NaN']))
    
    return  x / y

P_Bc = count_Px('C')
P_Bq = count_Px('Q')
P_Bs = count_Px('S')

def count_PAx(embarked):
    x = count_Px(embarked) 
    y = float(len(survived_df[survived_df.Embarked==embarked])) / float(len(survived_df))
              
    return x * y

P_A_and_Bc = count_PAx('C')
P_A_and_Bq = count_PAx('Q')
P_A_and_Bs = count_PAx('S')
```

```python
P_ABc = P_A_and_Bc / P_Bc
P_ABq = P_A_and_Bq / P_Bq
P_ABs = P_A_and_Bs / P_Bs

print P_ABc, P_ABq, P_ABs
```
